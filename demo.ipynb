```python
processed_transactions = transactions.copy()

### To give more details of what the transactions are ### 
processed_transactions = processed_transactions.merge(articles[['article_id','product_type_name', 'product_group_name', 'graphical_appearance_name', 'colour_group_name', 'perceived_colour_value_name', 'index_group_name', 'section_name']], how='left', on='article_id')
processed_transactions
```

```python
processed_customers = customers.copy()

### Transaction count per customer ###
processed_customers = processed_customers.merge(transactions.groupby(['customer_id'])['article_id'].count(), how='left', on='customer_id')
processed_customers.rename(mapper={"article_id":"Overall_transaction_count"}, axis='columns', inplace=True)

### Num of transactions through sales channel ID per customer ###
processed_customers = processed_customers.merge(transactions[transactions['sales_channel_id'] == 1].groupby(['customer_id'])['article_id'].count(), how='left', on='customer_id')
processed_customers.rename(mapper={"article_id":"Overall_transaction_count_channel_1"}, axis='columns', inplace=True)
processed_customers = processed_customers.merge(transactions[transactions['sales_channel_id'] == 2].groupby(['customer_id'])['article_id'].count(), how='left', on='customer_id')
processed_customers.rename(mapper={"article_id":"Overall_transaction_count_channel_2"}, axis='columns', inplace=True)

### Last transaction date ###
processed_customers = processed_customers.merge(transactions.groupby('customer_id')['t_dat'].max(), how='left', on='customer_id')
processed_customers.rename(mapper={"t_dat":"Last_transaction_date"}, axis='columns', inplace=True)

### Total amount bought through the period ###
processed_customers = processed_customers.merge(transactions.groupby('customer_id')['price'].sum(), how='left', on='customer_id')
processed_customers.rename(mapper={"price":"Overall_transacted"}, axis='columns', inplace=True)
```

```python
### Note that these codes will take 2.5mins each to run.
### Product_type bought most ###
processed_customers = processed_customers.merge(processed_transactions.groupby(['customer_id', 'product_type_name'])['price'].sum().groupby('customer_id').idxmax().apply(lambda x: x[1]), how='left', on='customer_id')
processed_customers.rename(mapper={"price":"Product_type_bought_most"}, axis='columns', inplace=True)

### Graphical_appearance bought most ###
processed_customers = processed_customers.merge(processed_transactions.groupby(['customer_id', 'graphical_appearance_name'])['price'].sum().groupby('customer_id').idxmax().apply(lambda x: x[1]), how='left', on='customer_id')
processed_customers.rename(mapper={"price":"Graphic_bought_most"}, axis='columns', inplace=True)

### Perceived color bought most ###
processed_customers = processed_customers.merge(processed_transactions.groupby(['customer_id', 'perceived_colour_value_name'])['price'].sum().groupby('customer_id').idxmax().apply(lambda x: x[1]), how='left', on='customer_id')
processed_customers.rename(mapper={"price":"Perceived_color_bought_most"}, axis='columns', inplace=True)

### Section_name bought most ###
processed_customers = processed_customers.merge(processed_transactions.groupby(['customer_id', 'section_name'])['price'].sum().groupby('customer_id').idxmax().apply(lambda x: x[1]), how='left', on='customer_id')
processed_customers.rename(mapper={"price":"Section_bought_most"}, axis='columns', inplace=True)
```

```python
processed_customers_adrian= processed_customers.copy()

processed_customers_adrian = processed_customers_adrian.merge(transactions.groupby('customer_id')['t_dat'].min(), how='left', on='customer_id')
processed_customers_adrian.rename(mapper={"t_dat":"Earliest_transaction_date"}, axis='columns', inplace=True)

latest_date = transactions['t_dat'].max()

processed_customers_adrian['Recency'] = latest_date - processed_customers_adrian['Last_transaction_date']
processed_customers_adrian['Tenure'] = processed_customers_adrian['Last_transaction_date'] - processed_customers_adrian['Earliest_transaction_date']
```

```python
processed_customers_adrian = processed_customers_adrian.merge(transactions.groupby('customer_id')['article_id'].nunique(), how='left', on='customer_id')
processed_customers_adrian.rename(mapper={"article_id":"Transaction_count_revised"}, axis='columns', inplace=True)

processed_customers_adrian
```

```python
processed_customers_adrian['Recency_days'] = processed_customers_adrian['Recency'].dt.days
processed_customers_adrian['Tenure_days'] = processed_customers_adrian['Tenure'].dt.days

processed_customers_adrian = processed_customers_adrian.merge(transactions.groupby('customer_id')['article_id'].nunique(), how='left', on='customer_id')
processed_customers_adrian.rename(mapper={"article_id":"Unique_items"}, axis='columns', inplace=True)
processed_customers_adrian['Unique_items'].fillna(0, inplace=True)

```

```python
#Customer Total Spend
transactions_b = transactions.copy()
transactions_b['total_spent'] = transactions_b.groupby('customer_id')['price'].transform('sum')

#Seasonality
transactions_b['t_dat'] = pd.to_datetime(transactions_b['t_dat'])

# Extract month, day of week, and week of year from t_dat
transactions_b['month'] = transactions_b['t_dat'].dt.month
transactions_b['day_of_week'] = transactions_b['t_dat'].dt.dayofweek
transactions_b['week_of_year'] = transactions_b['t_dat'].dt.isocalendar().week

# Unique products
transactions_b['unique_products'] = transactions_b.groupby('customer_id')['article_id'].transform('nunique')


transactions_b.head()
```

```python
# Bin Age
customers_b = customers.copy()
bins = [18, 24, 34, 44, 54, 64, 100]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
customers_b['age_group'] = pd.cut(customers_b['age'], bins=bins, labels=labels, right=False)

def categorize_age(age):
    if age < 25:
        return 'Young'
    elif age < 50:
        return 'Adult'
    else:
        return 'Senior'

# Apply the function to the 'age' column
customers_b['age_group'] = customers_b['age'].apply(categorize_age)


customers_b.head()
```

```python
# Monthly Transactions
sns.countplot(x='month', data=transactions_b)
plt.title('Number of Transactions by Month')
plt.show()

# Transactions by Day of Week
sns.countplot(x='day_of_week', data=transactions_b)
plt.title('Number of Transactions by Day of Week')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(14,10))

# Club Member Status
sns.countplot(ax=ax[0,0], x='club_member_status', data=customers_b)
ax[0,0].set_title('Club Member Status')

# Fashion News Frequency
sns.countplot(ax=ax[0,1], x='fashion_news_frequency', data=customers_b)
ax[0,1].set_title('Frequency of Fashion News')

# Age Distribution
sns.histplot(ax=ax[1,0], x='age', data=customers_b, kde=True, bins=30)
ax[1,0].set_title('Age Distribution')

# Age Group
sns.countplot(ax=ax[1,1], x='age_group', data=customers_b)
ax[1,1].set_title('Distribution of Age Groups')

plt.tight_layout()
plt.show()
```
